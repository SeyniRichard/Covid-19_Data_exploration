```python
#Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#import data

df = pd.read_csv(r'C:\Users\sanos\Downloads\movies.csv')

df_original = df
```

```python
df_original.head()
```

```python
for col in df.columns:
    pct_missing =100* np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#check data types

df.dtypes
```

```python
df['budget'].fillna(df['budget'].mean(), inplace=True)
df['gross'].fillna(df['gross'].mean(), inplace=True)
```

```python
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Remove duplicates

df.drop_duplicates()
df.head()
```

```python
#checking correlations between variables

#scatter plot

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget and Revenues')

plt.xlabel('Budget')
plt.ylabel('Revenues')
plt.show()
```

```python
#With the normal scatter plot, it's hard to deduct the correlation strength but with regplot we may be able to get more clarity.

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

From here, we can clearly see that there is a positive correlation between buget of a movies and its revenues.

```python
df.corr()
```

Now, we can see that the correlation between budget and revenues(gross) is 0.71 which is pretty high

```python
# The Heatmap can better show the strength of each correlation

corr_matrix =df.corr()

sns.heatmap(corr_matrix, annot=True)

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df_original.head()
```

```python
corr_matrix =df_numerized.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation matrix for all numeric features')

plt.show()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs= correlation_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr= sorted_pairs[(sorted_pairs)>0.4]

high_corr
```

Conclusion:  Budget and Votes have the highest correlation to gross revenues of a movie.
